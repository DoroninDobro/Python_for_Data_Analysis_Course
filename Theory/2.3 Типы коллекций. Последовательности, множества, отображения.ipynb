### Последовательности

Последовательности - это один из типов данных, поддерживающих оператор проверки на вхождение (```in```), функцию определения размера (```lеn()```), оператор извлечения срезов (```[]```) и возможность выполнения итераций. 

#### Кортежи

Кортеж - это неизменяемая упорядоченная последовательность из нуля и более ссылок на объекты.

````
# Создание кортежей
tp1 = tuple()
tp2 = (1,2,3)

# Количество элементов в кортеже
print(len(tp1))
print(len(tp2))

print(type(tp1))
print(type(tp2))
````

````
# Позиции элементов в кортеже
print(tp2[0], tp2[1], tp2[2])
print(tp2[-3], tp2[-2], tp2[-1])
````

````
# Операция среза в кортеже
print(tp2[:])
print(tp2[-3:])
print(tp2[0:-1])
print(tp2[0:2])
print(tp2[::2])
print(tp2[::-1])
````

````
# Проверка неизменяемости
tp2[0]=1
````

````
#  Работа с функциями
def f_sum_prod(a, b):
  return a + b, a * b

f_sum_prod(2, 5)
````

````
# Использование функции zip
for i in zip((1,2),(3,4)):
  print(i)
````

#### Именованные кортежи

````
import collections
dir(collections)
````

````
# Создание прототипа именнованного кортежа
User = collections.namedtuple('User','name email reg_date')
````

````
# Кортеж из пользователей
users_list = (User('Boris','b@gmail.com','01.02.00'),
              User('Anton','a@gmail.com','11.02.01'),
              User('Don','d@gmail.com','01.04.02'))
````

````
users_list
````

````
# Обращение к записям
print(users_list[0].name)
print(users_list[0].reg_date)
````

````
# Пример удобства в чтении кода
Iam = User('Alex','al@gmail.com','01.02.02')
print(Iam.name, Iam[0])
print(Iam.email, Iam[1])
print(Iam.reg_date, Iam[2])
````

#### Списки

Список - это изменяемая упорядоченная последовательность из нуля и более ссылок на объекты.

Методы списков:
![alt text](https://drive.google.com/uc?id=1_HbvqT7omF7rxmtZ3PHLzeoeqxxccVM2)

````
# Создание списков
list1 = list()
list2 = []
list3 = [1,2,3,4]
````

````
# Добавление элементов в списки
list1.append(1)
print(list1)
````

````
# Добавление списка в качестве элемента списка
list1.append(list3)
print(list1)
````

````
# Конкатенирование двух списков
list1 += list3
print(list1)
````

````
# Индексация списка
print(list1[0])
print(list1[1])
print(list1[5])
print(list1[1][2])
````

````
# Операция среза
print(list1[:])
print(list1[1:])
print(list1[1][2:])
print(list1[::-1])
````

````
# Удаление элементов списка
list1.remove(2)
print(list1)
````

````
del list1[0]
print(list1)
````

````
del list1[0][3]
print(list1)
````

````
# Вставка элемента
list1.insert(0, list3)
print(list1)
````

````
# В list3 косвенно был удален элемент, поскольку list1[0][3] - это ссылка на элемент в list3
list3
````

#### Множества

Будем рассматривать только тип ```set``` - неупорядоченную коллекцию из нуля и более ссылок различных объектов. Поддерживает операции на множествах (в математическом смысле): объединение, пересечение, разность и т.д.

````
# Создание множества
st1 = set()
st2 = {}
st3 = {1,2,3,3,4,2}
st4 = {1,4,6,8}
````

````
# Пустые {} создают не множество set!
print(type(st1))
print(type(st2))
````

````
# Одно из возможных использований - получение множества уникальных значений
print(st3)
````

````
# Добавление элементов в множество
st3.add(0)
print(st3)
````

````
st4
````

````
# Пересечение множеств
st3 & st4
````

````
# Объединение множеств
st3 | st4
````

````
# Разность множеств
st3 - st4
````

````
# Отсутствие упорядоченности
st3[0]
````

### Отображения

#### Словари

Тип ```diсt``` -
это неупорядоченная коллекция из нуля или более пар
«ключ-значение», в которых в качестве ключей могут использоваться
ссылки на неизменяемые (хешируемые)объекты, а в качестве значений -
ссылки на объекты любого типа.

![alt text](https://drive.google.com/uc?id=1Wy8CdeDiKaPu1wKLDT7rYdOGv24lteVi)

````
# Создание словаря
dict1 = {}
dict2 = dict()
dict3 = {'a':1, 'b':2, 4:[1,2], (1,'Ted'):[(1,2),1,'abc']}
dict4 = dict.fromkeys(['a','b'],0)
dict5 = dict(zip(['a','b'],[0,1]))
````

````
print(dict3)
print(dict4)
print(dict5)
````

````
# Добавление значений
dict5.update({'c':3})
print(dict5)
````

````
# Индексация
print(dict3['a'])
print(dict3['b'])
print(dict3[(1,'Ted')])
print(dict3[4])
````

````
# Проверка упорядоченности
print(dict3[0])
````

````
# Просмотр всех значений, ключей, пар ключ-значение
print(dict3.values())
print(dict3.keys())
print(dict3.items())
````
