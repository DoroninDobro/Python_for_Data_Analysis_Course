## Задание 1

**Условие:** 
В учебных целях часто прибегают к использованию готовых датасетов, как например датасет [iris](https://ru.wikipedia.org/wiki/%D0%98%D1%80%D0%B8%D1%81%D1%8B_%D0%A4%D0%B8%D1%88%D0%B5%D1%80%D0%B0).

Код ниже позволяет скачать указанный датасет в формате: 

*   ```X``` - массив объектов, в каждом по 4ре значения признаков в соответствии с ```feature_names```
*   ```y``` - массив меток объеков в соответствии с ```target_names``` (всего 3 возможных значения)

Требуется выполнить следующие действия:


*   нарисовать точечные диаграммы (```scatter plot```) всех четырех признаков, получив значения из соответствующих столбцов ```X```. Необходимо для этого использовать ```subplot``` формата ```2x2```.
*   добавить названия признаков к каждому графику в качестве имени графика
*   использовать метки классов ```y``` в качестве цветовой разметки (```plt.scatter(x, y, color)```)
*   указать название признака, для которого на ваш взгляд множество точек наиболее хорошо сгруппировалось по цветам


**Формат ответа:** 
Строка (например, ```'sepal length (cm)'```)

**Оценка в баллах:** 3 балла

**Решение:**

````
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data = datasets.load_iris(return_X_y=False)
X = data.data
y = data.target
feature_names = data.feature_names
target_names = data.target_names
````

````
data.feature_names
````

````
fig, ax = plt.subplots(2, 2, figsize=(10, 10))
for i in range(4):
  plt.subplot(221 + i)   
  plt.scatter(range(X.shape[0]), X[:, i], c = y)
  plt.title(feature_names[i])
fig.show()
````

**Ответ:**
