````
import seaborn as sns
sns.set(style="ticks")
sns.set(rc={'figure.figsize':(8,8)})
data = sns.load_dataset("planets")
data.dropna(inplace=True)
data.head()
data.info()
````

````
sns.relplot(x="distance", y="mass",data=data, hue=data.method)
````

````
sns.countplot(x="number", data=data)
````

````
sns.barplot(x="number", y="mass", hue="method", data=data)
````

````
sns.boxplot(x="year", y="number", orient="h", data=data)
````

````
sns.violinplot(x="year", y="number", orient="h", data=data)
````

````
sns.distplot(data.distance,rug=True)
````

````
sns.lmplot(x="mass", y="distance", data=data, hue='number')
````

````
sns.lmplot(x="mass", y="distance", col='number', data=data)
````

````
sns.lmplot(x="mass", y="distance", data=data[data['number']==2])
````

````
sns.pairplot(data)
````

````
sns.heatmap(data.groupby(['number','year'])['distance'].mean().unstack().fillna(0))
````
