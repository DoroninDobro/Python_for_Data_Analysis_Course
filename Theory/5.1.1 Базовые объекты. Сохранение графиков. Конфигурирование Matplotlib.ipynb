## Материал

### Быстрый старт с графиками по умолчанию

````
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

data = np.arange(10)
print('Постройка графика функции со значениями:',data)
plt.plot(data)
````

````
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
````

### Цвета, маркеры и стили линий

````
print('Поточеченый график в виде красных точек:')
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
````

````
print('Различные типы и цвета отрисовки линий:')
t = np.arange(0., 5., 0.2)

plt.figure()
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'bs')
plt.plot(t, t**3, 'g^')

plt.plot(t, t**3.2, c='g', ls='steps')

plt.plot(t, t**3.4, c='g', ls='--', marker='+')
plt.show()
````

````
print('Толщина линий:')
plt.plot(t, t, linewidth=2)
plt.plot(t, t**2, linewidth=4)
plt.plot(t, t**3, linewidth=8)
plt.show()
````

### Рисунки и подграфики

````
print('Использование объекта figure():')
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo')
plt.plot(t2, t2**2*f(t2), 'b--')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.show()
````

````
print('Пример альтернативного кода:')
fig, axes = plt.subplots(2,1)

axes[0].plot(t1, f(t1), 'bo')
axes[0].plot(t2, t2**2*f(t2), 'b--')

axes[1].plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.show()
````

````
print('Очистка объекта fig:')
fig.clear()
fig
````

````
print('Более объемный пример:')
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

f, axes = plt.subplots(2, 2, sharex=False, sharey=False)

axes[0,0].plot(x,y)
axes[0,1].plot(x, np.abs(y))
axes[1,0].plot(x,y**3)
axes[1,1].plot(x,np.arccos(y))

plt.show()
````

### Задание названия графика и координатных осей

````
print('Различные типы и цвета отрисовки линий:')
t = np.arange(0., 5., 0.2)

plt.figure()
plt.title('Различные показательные функции')

plt.plot(t, t, 'r--',label='linear')
plt.plot(t, t**2, 'bs',label='x^2')
plt.plot(t, t**3, 'g^',label='x^3')

plt.xlabel('Абсцисса')
plt.ylabel('Ордината')
plt.legend()
plt.show()
````

### Сохранение графиков в файл

````
t = np.arange(0., 5., 0.2)

plt.title('Различные показательные функции')

plt.plot(t, t, 'r--',label='linear')
plt.plot(t, t**2, 'bs',label='x^2')
plt.plot(t, t**3, 'g^',label='x^3')

plt.xlabel('Абсцисса')
plt.ylabel('Ордината')
plt.legend()
plt.grid()
plt.savefig('ThreeGraphs.png')
````
