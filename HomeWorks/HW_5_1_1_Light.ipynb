## Задание 1

**Условие:** 
Постройте график функции $y=2*\sin x + 4$. Подпишите оси. Дайте название графику. Укажите в ответе минимальное и максимальное значения для данный функции.

**Формат ответа:** 
Кортеж чисел (например, ```(0,5)```)

**Оценка в баллах:** 3 балла

**Решение:**

````
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline

x = np.arange(-100,100, 0.5)
y=2*np.sin(x)+4
plt.figure()
plt.plot(x,y)
plt.title('Task 1')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
ans1 = (y.min(), y.max())
print(ans1)
````

## Задание 2

**Условие:** 
Постройте графики функций $y=x^2$ и $y=x$ на одном графике. Проделайте следующие операции:

* дайте название графику
* дайте названия осям
* присвойте лейблы и выведите легенду
* отрисуйте графики функций пунктирными линиями и разными цветами

Укажите максимальное значение среди абсцисс точек пересечения графиков, округленных до целого.

**Формат ответа:** 
Число (например, ```3```)

**Оценка в баллах:** 3 балла

**Решение:**

````
x = np.arange(-2,2,0.2)
y1 = x**2
y2 = x

plt.figure(figsize=[10,5])
plt.title('График')
plt.plot(x,y1,label='y=x^2', c='r',ls='--')
plt.plot(x,y2,label='y=x', c='g',ls='--')
plt.axis([x.min()-1,x.max()+1,y1.min()-1,y1.max()+1])
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
i = 0
ans2 = x.min()
for i in range (len(y1)):
  ans2 = round(x[i],0) if round(y1[i]-y2[i],2)==0 else ans2

print(ans2)
````

**Ответ:**

````
# Не изменяйте текст до знака = во избежании ошибок проверки!
_HOMEWORK_ANSWER_DICT_['A2'] = ans2
````
