### Диаграмма




Дополнительные [примеры](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.bar.html#matplotlib.pyplot.bar) кода.

````
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(5)
height = [20, 23, 25, 24, 26]

plt.bar(x, height, align='center', color='c', linewidth = 100)

plt.title('Продажи товара Х по месяцам')

plt.ylabel('млн руб.')
plt.xlabel('Месяц')
plt.xticks(x, ('Январь', 'Февраль', 'Март', 'Апрель', 'Май'))

plt.show()
````

### Гистограмма





Дополнительные [примеры](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html) кода.

````
np.random.seed(42)
x = np.random.normal(0, 1, 100000)

plt.hist(x, bins=10)
plt.title('Гистограмма распределения значений в массиве х')
plt.show()
````

### Контурные графики





Дополнительные [примеры](https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.contour.html) кода.

````
def f(x,y):
    return x**2 + y**2
````

````
x = np.linspace(-10, 10, 1000)
y = np.linspace(-10, 10, 1000)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
````

````
plt.figure(figsize=(5,5))
plt.contour(X, Y, Z, 20)
plt.show()
````

````
plt.figure(figsize=(5,5))
plt.contourf(X, Y, Z,20)
plt.show()
````

### Тепловая карта (Heatmap)

Дополнительные [примеры](https://matplotlib.org/gallery/images_contours_and_fields/image_annotated_heatmap.html) кода.



````
np.random.seed(42)
corr = np.random.rand(10,10)
print(corr)
plt.imshow(corr)
plt.colorbar()
plt.show()
````

````
plt.imshow(corr, cmap='hot')
plt.colorbar()
plt.show()
````

````
print('Пример тепловой визуализации сбора урожая различных овощей фермерами:')

vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"]
farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]

harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])


fig, ax = plt.subplots(figsize=(7,7))
im = ax.imshow(harvest)

# Отрисовываем штрихи
ax.set_xticks(np.arange(len(farmers)))
ax.set_yticks(np.arange(len(vegetables)))

# Наносим подписи к штрихам
ax.set_xticklabels(farmers)
ax.set_yticklabels(vegetables)

# Поворачиваем вертикальные подписи для читабельности
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Подписываем каждую ячейку соответствующим значением
for i in range(len(vegetables)):
    for j in range(len(farmers)):
        text = ax.text(j, i, harvest[i, j],
                       ha="center", va="center", color="w")

ax.set_title("Harvest of local farmers (in tons/year)")

ax.figure.colorbar(im, ax=ax)

fig.tight_layout()
plt.show()
````

### Точечная диаграмма (Scatter plot)

````
N = 10
x = np.random.rand(N)
y = np.random.rand(N)
z = x*y
colors = z > 0.2

fig, ax = plt.subplots(2,1, figsize=(6,6))

ax[0].scatter(x, y, s=100*z, c=colors)
ax[0].set_title('Scatter plot')

ax[1].plot(x, y, 'yo')
ax[1].set_title('Basic plot with yellow circles')

fig.tight_layout()
plt.show()
````
