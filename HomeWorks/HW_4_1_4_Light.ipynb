## Задание 3

**Условие:** 
Создайте новый столбец ```TotalCost``` со значениями ```Quantity * UnitPrice```. Сделайте ```resampling``` данных в еженедельные со значениями суммы по столбцам ```TotalCost```. Найдите максимальное значение в столбце ```TotalCost```.

**Формат ответа:** 
Число (например, ```'12362'```)

**Оценка в баллах:** 3 балла

**Решение:**

````
data['TotalCost'] = data.Quantity * data.UnitPrice
ans3 = round(data.resample('W').sum().TotalCost.max(), 2)
print(ans3)
````

**Ответ:**

````
# Не изменяйте текст до знака = во избежании ошибок проверки!
_HOMEWORK_ANSWER_DICT_['A3'] = ans3
````
