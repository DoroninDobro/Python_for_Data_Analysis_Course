## Материал

### Условные операторы

Общий вид условного ветвления в языке Python:


```
if boolean_expression1:
  suitel
elif boolean_expression2:
  suite2
...
elif boolean_expressionN:
  suiteN
else:
  else_suite
```
Данная конструкция может содержать ```0``` и более ответвлений ```elif```. Конструкция ```else``` также является необязательной.

```if``` - если произойдет...

```elif``` - иначе в случае...

```else``` - если ни один альтернативный случай не учтен...




````
# Простой пример
a = int(input('Enter an integer number:'))

if a > 0:
  print('a > 0')
else:
  print('a <= 0')
````

````
# Пример ветвления
a = int(input('Enter an integer number:'))

if a < 10:
  print('a < 10')
elif a < 15:
  print('10 <= a < 15')
elif a < 20:
  print('15 <= a < 20')
else:
  print('a >= 20')
````

````
# Пример важности порядка ветвления
a = int(input('Enter an integer number:'))

if a < 10:
  print('a < 10')
elif a < 20:
  print('15 <= a < 20')
elif a < 15:
  print('10 <= a < 15')
else:
  print('a >= 20')
````

````
# Несколько условий
a = int(input('Enter an integer number:'))

if a%2==0 and a%3==0:
  print('6|a')
elif a%2==0 and a%3!=0:
  print('2|a')
elif a%2!=0 and a%3==0:
  print('3|a')
else:
  print('2,3,6 not | a')
````

````
# Короткая версия условного ветвления
a = int(input('Enter an integer number:'))
print('2 not | a') if a%2 == 1 else print('2|a')
````

### Циклы


Общая конструкция цикла ```while```:


```
while boolean_expression:
  while_suite
else:
  else_suite
```
Предложение ```else``` является необязательным - выполняется, когда выражение boolean_expression становится ложным. Если же выполнение прерывается инструкциями ```return``` или ```break```, то блок ```else``` не выполняется.



````
# Поиск элемент в list
my_list=[1,2,'c',4,'a']

target = 4

index = 0
while index < len(my_list):
  print(index)
  if my_list[index] == target:
    break
  else:
    index += 1
else:
  index = -1

print(index)
````

````
# Пример без else - вывод чисел от 1 до 35 кратных 5
i = 1
while i <= 35:
  if not i%5:
    print(i, end=' ')
  i += 1
````

Общая конструкция цикла ```for ... in ...```


```
for expression in iterable:
  for_suite
else:
  else_suite 
```

```expression``` - переменная или последовательность переменных

```iterable``` - итерируемый объект



````
# Простой пример 
for i in my_list:
  print(i, end=' ')

print()

for i in range(len(my_list)):
  print(my_list[i], end=' ')
````

````
# Выведем еще и индексы
for index, i in enumerate(my_list):
  print(index, i)
````

````
# Наиболее распространенный способ применения цикла for
for i in range(10):
  print(i, end=' ')
````

````
# Обходим наш лист, пропуская элемент с индексом 2
for i in range(len(my_list)):
  if i == 2:
    continue
  print(i, my_list[i])
````

````
# Обходим наш лист до элемента с индексом 2
for i in range(len(my_list)):
  if i == 2:
    break
  print(i, my_list[i])
````
