Об использовании всего арсенала функций библиотеки ```seaborn``` на примере анализа конкретного датасета ```mpg.csv```.

## Ключевые слова, понятия и выражения






*   ```seaborn```

## Материал

````
import seaborn as sns
sns.set(style="ticks")
sns.set(rc={'figure.figsize':(8,8)})
````

````
data = sns.load_dataset("mpg")
data.dropna(inplace=True)
data.head()
````

````
data.info()
````

````
sns.relplot(x="horsepower", y="mpg",data=data)
````

````
sns.countplot(x="origin", data=data)
````

````
sns.barplot(x="model_year", y="mpg", hue="origin", data=data)
````

````
sns.boxplot(x="mpg", y="model_year", orient="h", data=data)
````

````
sns.violinplot(x="mpg", y="model_year", orient="h", data=data)
````

````
sns.distplot(data['mpg'],rug=True)
````

````
sns.distplot(data['weight'],rug=True)
````

````
sns.lmplot(x="weight", y="mpg", data=data)
````

````
sns.lmplot(x="weight", y="mpg", hue='origin', data=data)
````

````
sns.lmplot(x="weight", y="mpg", col='origin', data=data)
````

````
sns.lmplot(x="weight", y="mpg", data=data[data['origin']=='japan'])
````

````
sns.pairplot(data.drop(columns=['origin', 'name']))
````

````
data.groupby(['origin','cylinders'])['mpg'].mean().unstack().fillna(0)
````

````
sns.heatmap(data.groupby(['origin','cylinders'])['mpg'].mean().unstack().fillna(0))
````

````
sns.heatmap(data.groupby(['model_year','cylinders'])['mpg'].mean().unstack().fillna(0))
````
