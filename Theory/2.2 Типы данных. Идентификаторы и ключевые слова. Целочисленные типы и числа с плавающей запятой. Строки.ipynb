### Идентификаторы и ключевые слова

Идентификаторы - имена, которые даются ссылкам на объекты. Общая договоренность по использованию идентификаторов дается в [PEP8](https://www.python.org/dev/peps/pep-0008/#naming-conventions).

Cтили идентификаторов:
*    b (одиночная маленькая буква)
*    B (одиночная заглавная буква)
*    lowercase (слово в нижнем регистре)
*    lower_case_with_underscores (слова из маленьких букв с подчеркиваниями)
*    UPPERCASE (заглавные буквы)
*    UPPERCASE_WITH_UNDERSCORES (слова из заглавных букв с подчеркиваниями)
*    CapitalizedWords (слова с заглавными буквами, или CapWords, или CamelCase)
*    mixedCase (отличается от CapitalizedWords тем, что первое слово начинается с маленькой буквы)
*    Capitalized_Words_With_Underscores (слова с заглавными буквами и подчеркиваниями — уродливо!)

Ключевые слова - зарезервированные языком программирования слова, которые не должны присваиваться ссылкам на объекты во избежании ошибок.

### Комплексные числа

````
z = -4.0 + 3.0j
print(type(z))
print('Re=',z.real)
print('Im=',z.imag)
````

### Строки

````
# Использование кавычек
str1 = "It's a table"
print(str1)
str1 = "It is \"a table\""
print(str1)
````

````
# Использование апострофов
str2 = 'It is "a table"'
print(str2)
str2 = 'It\'s a table'
print(str2)
````

````
# Использование тройных кавычек
str3 = '''- Привет! 
- Как "дела'?
- Хорошо'''
print(str3)
````

````
str3
````

````
# Срезы строк
print(str1)
print('-'*10)
print(str1[0:5])
print('-'*10)
print(str1[3:5])
print('-'*10)
print(str3)
print('-'*10)
print(str3[-6:])
print('-'*10)
print(str3[0::1])
print('-'*10)
print(str3[0::2])
print('-'*10)
print(str3[::-1])
````
